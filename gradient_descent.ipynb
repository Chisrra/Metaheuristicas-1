```python
import numpy as np
import matplotlib.pyplot as plt
```

## Parametros

```python
LEARNING_RATE = 0.01
NUM_ITERATIONS = 100
INITICIAL_X = np.random.uniform(-5,5)
```

## Funciones

```python
def f(x:float):
    return 2*(x**2)*np.cos(x) - 5*x

def df(x:float):
    return 4 * x * np.cos(x) - 2 * (x**2) * np.sin(x) - 5
```

## Implementacion

```python
x_values = []
fx_values = []
x = INITICIAL_X
```

```python
for _ in range(NUM_ITERATIONS):
    x_values.append(x)
    fx_values.append(f(x))
    x -= LEARNING_RATE * df(x)

print("El valor mínimo aproximado de la función: ", x)
print("Valor mínimo de la función: ", 3.7829803535708)
```

## Grafico

```python
x_range = np.linspace(-5, 5, 1000)
plt.plot( x_range, f(x_range), label = "Function $2x^2 \cdot \cos(x) - 5x$" )
plt.scatter( x_values, fx_values, color="blue", label="Convergencia del gradiente descendente" )
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.title("Gradiente Decendente para $2x^2 \cdot \cos(x) - 5x$")
plt.show()
```
