```python
import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Parametros
LERNING_RATE = 0.01
NUM_ITERATIONS = 2000
INITICIAL_X = random.uniform(-20,5)
INITICIAL_Y = random.uniform(-5,5)
```

```python
def f(x:float, y:float):
    return  x**2 + 3 * y

def dfx(x:float):
    return 2 * x

def dfy(y:float):
    return 3
```

```python
x_values = []
y_values = []
fxy_values = []
x = INITICIAL_X
y = INITICIAL_Y
```

```python
for _ in range(NUM_ITERATIONS):
    x_values.append(x)
    y_values.append(y)
    fxy_values.append( f(x, y) )
    x = x - LERNING_RATE * dfx(x)
    y = y - LERNING_RATE * dfy(y)

print(f"El valor mínimo aproximado de la función.\n x: {x}, y:{y} ")
print("El mínimo real de la función es: ❌")
```

```python
# Crear una malla de puntos para graficar
x_range = np.linspace(-40, 40, 100)
y_range = np.linspace(-40, 40, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

# Crear la figura y el eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie de la función
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)

# Graficar la trayectoria del descenso de gradiente
ax.plot(x_values, y_values, fxy_values, color='red', marker='o')

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title('Gradient Descent 3D: x^2+3y')

plt.show()
```
